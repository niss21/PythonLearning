```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(url,names = headers)
df.head()
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print(avg_norm_loss)
```

```python
# df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
df.replace({"normalized-losses": np.nan}, avg_norm_loss, inplace=True)
df.head()
```

```python
# df["bore"].replace(np.nan,df['bore'].astype('float').mean(axis=0),inplace=True)
# df.replace({"bore": {np.nan: df['bore'].astype('float').mean(axis=0)}}, inplace=True)
df["bore"] = df["bore"].replace(np.nan, df['bore'].astype('float').mean(axis=0))
df.head()
```

```python
df.replace({
    "stroke": {np.nan: df['stroke'].astype('float').mean(axis=0)},
    "horsepower": {np.nan: df['horsepower'].astype('float').mean(axis=0)},
    "peak-rpm": {np.nan: df['peak-rpm'].astype('float').mean(axis=0)}
}, inplace=True)

df.head()
```

```python
# df.replace({'num-of-doors':np.nan},"four",inplace=True)
df.replace(to_replace = np.nan, value = "four")
```
