```python
df = df.drop(columns = ['Android_version','Inbuilt_memory','fast_charging','Screen_resolution','Processor'])
```

```python
df.shape
```

```python
df['Price'].str.replace(',', '').astype(int)
```

```python
df['Price'] = df['Price'].str.replace(',', '').astype(int)
```

```python
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df[['Spec_score']]
y = df[['Price']]
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(x,y)
```

```python
Yhat = lr.predict(x)
Yhat[0:5]
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.regplot(x="Spec_score", y="Price", data=df)
plt.ylim(0,)
```

```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.residplot(x=df['Spec_score'],y=df['Price'])
plt.show()
```
